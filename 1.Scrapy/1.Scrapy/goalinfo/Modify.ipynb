```python
import json
with open("Goal_info_2015.json","r") as f:
    goal = json.load(f)
```

```python
goal[1]
```

```python
goal_filter = dict()
for i in range(len(goal)):
    fill = []
    if(goal[i]['goal_information']['detail'] is not None):
        for g in goal[i]['goal_information']['detail']:
            name = goal[i]['goal_information']['id']
            if(g['season'] == '15/16' and
               g['league_name'] == "Premier League"):
                fill.append(g)
        goal_filter[name] = fill
```

```python
goal_filter['Memphis Depay']
```

```python
from collections import Counter
from collections import defaultdict
goal_provider = defaultdict(Counter)

for key in goal_filter.keys():
    if(len(goal_filter[key]) > 0):
        for g in goal_filter[key]:
            provider = g['provider_id']
            if(provider != 'null'):
                goal_provider[key][provider] = goal_provider[key].get(provider,0) + 1
        
```

```python
most_provider = dict()
for key in goal_provider.keys():
    most_provider[key] = goal_provider[key].most_common()[0]
```
